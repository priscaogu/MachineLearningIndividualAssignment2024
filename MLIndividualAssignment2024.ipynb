```python
def drivingStyle(x):
    if x == 'EvenPaceStyle':
        return 0
    else:
        return 1
        
#apply the function
opel_corsa_01['drivingStyle_new'] = opel_corsa_01['drivingStyle'].apply(drivingStyle)
peugeot_207_01['drivingStyle_new'] = peugeot_207_01['drivingStyle'].apply(drivingStyle)
```

```python
from scipy.stats import chi2_contingency, f_oneway


def anova_test(data, continuous_features, target_variable):
    anova_results = {}
    for feature in continuous_features:
        group_data = [group[1][feature].values for group in data.groupby(target_variable)]
        f_stat, p = f_oneway(*group_data)
        anova_results[feature] = {'F-statistic': f_stat, 'p_value': p}
    return anova_results

anova_results_op = anova_test(opel_corsa_01, continuous_features=['AltitudeVariation','VehicleSpeedInstantaneous','VehicleSpeedAverage','VehicleSpeedVariance','VehicleSpeedVariation','LongitudinalAcceleration','EngineLoad','EngineCoolantTemperature','ManifoldAbsolutePressure','EngineRPM','MassAirFlow','IntakeAirTemperature','VerticalAcceleration','FuelConsumptionAverage'], target_variable='drivingStyle_new')
anova_results_pu = anova_test(peugeot_207_01, continuous_features=['AltitudeVariation','VehicleSpeedInstantaneous','VehicleSpeedAverage','VehicleSpeedVariance','VehicleSpeedVariation','LongitudinalAcceleration','EngineLoad','EngineCoolantTemperature','ManifoldAbsolutePressure','EngineRPM','MassAirFlow','IntakeAirTemperature','VerticalAcceleration','FuelConsumptionAverage'], target_variable='drivingStyle_new')

print(anova_results_op)
print('--------------------------------------------------------------------------------------------------------------------------------------------')
print(anova_results_pu)
```

```python

from sklearn.preprocessing import StandardScaler


xpu = peugeot_207_01[['AltitudeVariation', 'VehicleSpeedInstantaneous', 'VehicleSpeedAverage', 'VehicleSpeedVariance', 'VehicleSpeedVariation', 'LongitudinalAcceleration',  'EngineCoolantTemperature', 'MassAirFlow',  'FuelConsumptionAverage']]
yypu = peugeot_207_01[['drivingStyle_new']]



# Step 2: Scale the features
scaler = StandardScaler()
Xpu_scaled = scaler.fit_transform(xpu)

# Use test_train_split to create the training and test sets
Xpu_train, Xpu_test, ypu_train, ypu_test = train_test_split(Xpu_scaled, yypu, test_size=.2, random_state=123)



print("Shape of Xpu_train:", Xpu_train.shape)
print("Shape of ypu_train:", ypu_train.shape)
```

```python
x = opel_corsa_01[['AltitudeVariation', 'VehicleSpeedInstantaneous', 'VehicleSpeedAverage', 'VehicleSpeedVariance', 'VehicleSpeedVariation', 'LongitudinalAcceleration',  'EngineCoolantTemperature', 'MassAirFlow',  'FuelConsumptionAverage']]
yy =opel_corsa_01[['drivingStyle_new']]

# Step 2: Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(x)

# Use test_train_split to create the training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, yy, test_size=.2, random_state=123)



print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
```

## BUILD THE MODEL

```python
#Build the model for opel corsa

from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

svm_model = SVC(kernel='linear', probability=True, random_state=123)  
svm_model.fit(X_train, y_train.values.ravel())  

logistic_model = LogisticRegression(random_state=123, max_iter=1000)  
logistic_model.fit(X_train, y_train.values.ravel())


#Build the model for Puegeot 201
svmpu_model = SVC(kernel='linear', probability=True, random_state=123)  
svmpu_model.fit(Xpu_train, ypu_train.values.ravel())  

logisticpu_model = LogisticRegression(random_state=123, max_iter=1000)  
logisticpu_model.fit(Xpu_train, ypu_train.values.ravel())  
```

```python
#make predictions for opel corsa

svm_pred = svm_model.predict(X_test)
svm_proba = svm_model.predict_proba(X_test)[:, 1]

lr_pred = logistic_model.predict(X_test)
lr_proba = logistic_model.predict_proba(X_test)[:, 1]


#make predictions for puegeot 201

svmpu_pred = svmpu_model.predict(Xpu_test)
svmpu_proba = svmpu_model.predict_proba(Xpu_test)[:, 1]

lrpu_pred = logisticpu_model.predict(Xpu_test)
lrpu_proba = logisticpu_model.predict_proba(Xpu_test)[:, 1]
```

## Validate the Models

```python
#Evaluate the model for opel corsa
import numpy as np
from sklearn.metrics import confusion_matrix, roc_auc_score, f1_score, accuracy_score, roc_curve, auc, classification_report
from sklearn.metrics import log_loss


print("SVM")

f1 = f1_score(y_test, svm_pred, average='macro')
conf_matrix = confusion_matrix(y_test, svm_pred)
roc_auc_svm = roc_auc_score(y_test,svm_proba)

print(classification_report(svm_pred, y_test))
print("Confusion Matrix:")
print(conf_matrix)

#Compute ROC curve
svm_fpr, svm_tpr, thresholds = roc_curve(y_test, svm_proba)

# Compute AUC
roc_auc_svm = auc(svm_fpr, svm_tpr)

print('\n')


print("LOGISTIC REGRESSION")

f1 = f1_score(y_test,lr_pred, average='macro')
conf_matrix = confusion_matrix(y_test, lr_pred)
print(classification_report(lr_pred, y_test))
print("Confusion Matrix:")
print(conf_matrix)

#Compute ROC curve
lr_fpr, lr_tpr, thresholds = roc_curve(y_test, lr_proba)

# Compute AUC
roc_auc_lr = auc(lr_fpr, lr_tpr)
print('\n')


# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(svm_fpr, svm_tpr, color='blue', lw=2, label=f'SVM (AUC = {roc_auc_svm:.4f})')
plt.plot(lr_fpr, lr_tpr, color='orange', lw=2, label=f'LR (AUC = {roc_auc_lr:.4f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
